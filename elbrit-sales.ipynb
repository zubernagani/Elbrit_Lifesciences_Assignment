```python

import numpy as np
import pandas as pd
```

```python
#Loading the excel files

df = pd.read_excel("/kaggle/input/eld-sales-data/ELB-Sales-Data.xlsx")
```

```python
# to understand the data and data structres display the first few or last few records

print(df.head())
```

```python
#Check the missing values

print(df.isnull().sum())
```

```python
# Fill missing values in 'HQ' and 'Sales Team' with 'Unknown'
df['HQ'].fillna('Unknown', inplace=True)
df['Sales Team'].fillna('Unknown', inplace=True)

df['Return for Reason'].fillna('Not specified', inplace=True)

# Fill missing values in 'Sales Invoice Name' with 'Unknown'
df['Sales Invoice Name'].fillna('Unknown', inplace=True)

# Check the dataframe again for any remaining missing values
print(df.isnull().sum())
```

**Question no 1 : What was the highest-selling product in September?**

```python
# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract the month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Filter data for September (month = 9)
september_data = df[df['Month'] == 9]

# Calculate sales for each product
september_sales = september_data.groupby('Item Name')['Primary Sales'].sum()

# Find the highest-selling product
highest_selling_product_september = september_sales.idxmax()
print(f"Highest-selling product in September: {highest_selling_product_september}")
```

**2. Which product had the highest sales for the "CND Chennai" sales team in May?**

```python
# Filter data for May and CND Chennai sales team
may_chennai_data = df[(df['Date'].dt.month == 5) & (df['Sales Team'] == 'CND Chennai')]

# Calculate sales for each product
may_chennai_sales = may_chennai_data.groupby('Item Name')['Primary Sales'].sum()

# Find the highest-selling product
highest_selling_product_may_chennai = may_chennai_sales.idxmax()
print(f"Highest-selling product in May for CND Chennai: {highest_selling_product_may_chennai}")
```

**3. Which customer had the maximum stock returns in October for the Bangalore HQ?**

```python
# Filter data for October and Bangalore HQ
october_bangalore_data = df[(df['Date'].dt.month == 10) & (df['HQ'] == 'Bangalore')]

# Check if the filtered DataFrame is empty
if october_bangalore_data.empty:
    print("No data available for Bangalore HQ in October.")
else:
    # Calculate stock returns for each customer
    october_bangalore_returns = october_bangalore_data.groupby('Customer')['Sales Return'].sum()

    # Find the customer with maximum stock returns
    if not october_bangalore_returns.empty:
        max_returns_customer_october_bangalore = october_bangalore_returns.idxmin()  # Use idxmin() for minimum value
        max_returns_value = october_bangalore_returns.min()

        print(f"Customer with maximum stock returns in October for Bangalore HQ: {max_returns_customer_october_bangalore}")
        print(f"Maximum stock return value: {max_returns_value}")
    else:
        print("No stock returns data available for Bangalore HQ in October.")
```

**4. Which sales team had the maximum percentage of primary sales returned due to 
expiry?**

```python
# Calculate the percentage of returns due to expiry for each sales team
df['Expiry Returns Percentage'] = (df['Against Expiry'] / df['Primary Sales']) * 100

# Find the sales team with the maximum percentage of returns due to expiry
max_expiry_returns_sales_team = df.groupby('Sales Team')['Expiry Returns Percentage'].mean().idxmax()
print(f"Sales team with maximum percentage of primary sales returned due to expiry: {max_expiry_returns_sales_team}")
```

**5. What percentage of overall primary sales was affected by breakage?**

```python
# Calculate the total primary sales and total breakage
total_primary_sales = df['Primary Sales'].sum()
total_breakage = df['Breakage'].sum()

# Calculate the percentage of primary sales affected by breakage
breakage_percentage = (total_breakage / total_primary_sales) * 100
print(f"Percentage of overall primary sales affected by breakage: {breakage_percentage:.2f}%")
```

**6. What were the primary sales for the Delhi HQ in the month of September?**

```python


# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter data for September and Delhi HQs
september_delhi_data = df[(df['Date'].dt.month == 9) & 
                          ((df['HQ'] == 'HQ- North Delhi') | 
                           (df['HQ'] == 'HQ-South Delhi') | 
                           (df['HQ'] == 'HQ-West Delhi') | 
                           (df['HQ'] == 'HQ-East Delhi'))]

# Calculate the total primary sales
september_delhi_primary_sales = september_delhi_data['Primary Sales'].sum()
print(f"Primary sales for Delhi HQ in September: {september_delhi_primary_sales}")
```

**7. What were the sales of "Britorva 20" in September for PALEPU PHARMA DIST PVT LTD 
under Coimbatore HQ?**

```python
# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter data for September, Coimbatore HQ, and specific customer and product
britorva_september_coimbatore = df[(df['Date'].dt.month == 9) & 
                                   (df['HQ'] == 'HQ-Coimbatore') & 
                                   (df['Customer'] == 'PALEPU PHARMA DIST PVT LTD CBE') & 
                                   (df['Item Name'] == 'BRITORVA 20')]

# Calculate the sales
britorva_september_sales = britorva_september_coimbatore['Primary Sales'].sum()
print(f"Sales of Britorva 20 in September for PALEPU PHARMA DIST PVT LTD under Coimbatore HQ: {britorva_september_sales}")
```

**Question 1 :What is the forecasted value of primary sales of october month**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np


# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate the primary sales by month
monthly_sales = df.resample('M', on='Date')['Primary Sales'].sum().reset_index()

# Prepare the data for linear regression
# Create a numeric representation of the month for the regression model
monthly_sales['Month_Num'] = monthly_sales['Date'].dt.month + (monthly_sales['Date'].dt.year - monthly_sales['Date'].dt.year.min()) * 12

# Use the month number as the feature and sales as the target
X = monthly_sales[['Month_Num']].values
y = monthly_sales['Primary Sales'].values

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Assuming the last month in your data is September and you want to predict for the next month
last_month_num = X[-1][0]
october_month_num = last_month_num + 1

# Forecast sales for October
october_sales_forecast = model.predict([[october_month_num]])
print(f"Forecasted Primary Sales for October: {october_sales_forecast[0]}")
```

```python
# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter data for CND Chennai sales team
chennai_data = df[df['Sales Team'] == 'CND Chennai']

# Aggregate the primary sales by month and product
monthly_sales = chennai_data.groupby([chennai_data['Date'].dt.to_period('M'), 'Item Name'])['Primary Sales'].sum().reset_index()
monthly_sales['Date'] = monthly_sales['Date'].dt.to_timestamp()

# Prepare the monthly numeric representation for regression
monthly_sales['Month_Num'] = monthly_sales['Date'].dt.month + (monthly_sales['Date'].dt.year - monthly_sales['Date'].dt.year.min()) * 12

# Initialize dictionary to store forecasted sales for each product
forecasted_sales = {}

# Train and forecast for each product
for product in monthly_sales['Item Name'].unique():
    product_sales = monthly_sales[monthly_sales['Item Name'] == product]
    X = product_sales[['Month_Num']].values
    y = product_sales['Primary Sales'].values
    
    # Train the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Forecast sales for November (assuming the last month in data is September)
    last_month_num = X[-1][0]
    november_month_num = last_month_num + 2  # October is +1, November is +2
    
    # Forecast sales for November
    forecasted_sales[product] = model.predict([[november_month_num]])[0]

# Find the product with the highest forecasted sales
highest_forecasted_product = max(forecasted_sales, key=forecasted_sales.get)
highest_forecasted_sales = forecasted_sales[highest_forecasted_product]

print(f"Product with the highest forecasted sales for CND Chennai in November: {highest_forecasted_product} with forecasted sales of {highest_forecasted_sales}")
```
